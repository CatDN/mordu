# Cubic VLE calculations vs experimental data, P-x diagrams

## vdW

```python
#temperatures to use: 298, 311, 378 (highest number of experimental points)

# classes
from eos import EOS
from eos_mixture import EOSMixture
from  alpha_r_cubic import AlphaRCubic
from mixture_fluid import MixtureFluid

# objects
from purefluid import NH3, H2
from cp0 import NH3_cp0_NIST, H2_cp0_NIST 


# algorithms and other functions and variables
from symbols import *
from mixture_rules import one_fluid_theory
from other_functions import multi_root
from algorithms import VLE_P_dew, VLE_P_bubble, vdW_coefficients

# general
import pandas as pd
import matplotlib.pyplot as plt

####################################### Temperature
temperatures = [298, 311, 378]

####################################### Experimental data
path ="../dataframes/ammonia_hydrogen_VLE.xlsx"
data = pd.read_excel(path)

# round temperature to closest kelvin
data["T"] = round(data["T"], 0)


####################################### EOS
#expressions for vdW
vdW = {
    "alpha_r_expr":  sp.log(abs(1/rho/(1/rho-b)))-a/(R*T*1/rho),
    "a_c_expr": 27/64*R**2*T_c**2/P_c,
    "alpha_T_expr": sp.sympify(1),
    "b_expr": 1/8*R*T_c/P_c
}

#create pure fluid EOS from PR
vdW_NH3 = EOS("vdW_NH3", NH3, NH3_cp0_NIST, AlphaRCubic, **vdW)
vdW_H2 = EOS("vdW_H2", H2, H2_cp0_NIST, AlphaRCubic, **vdW)

#create mixture and mixture EOS
mix = MixtureFluid(H2, NH3)

mix_EOS = EOSMixture("vdW_H2_NH3", mix, AlphaRCubic, vdW_H2, vdW_NH3, mixture_rule=one_fluid_theory)

####################################### calculations for various temperatures
for temperature in temperatures:
    # restrict data to matching temperature
    data_T = data[data["T"]==temperature]

    # calculate dew compositions and pressure from algorithm
    compositions_dew, pressures_dew = VLE_P_dew(mix_EOS, temperature, method="coefficients", pressure_break=400e5, step = 1e-2, coefficient_function=vdW_coefficients)

    # calculate bubble compositions and pressure from algorithm
    compositions_bubble, pressures_bubble = VLE_P_bubble(mix_EOS, temperature, method="coefficients", pressure_break=400e5, step = 0.005, pressure_jump=1e5, coefficient_function=vdW_coefficients)


    ####################################### Plotting
    fig, ax = plt.subplots()

    ax.plot(compositions_dew, pressures_dew, label="Dew line")
    ax.plot(compositions_bubble, pressures_bubble, label="Bubble line")

    ax.scatter(data_T["y_H2"], data_T["P"], label="Dew line experimental", color="black", marker="o")
    ax.scatter(data_T["x_H2"], data_T["P"], label="Bubble line experimental", color="black", marker="x")

    ax.set(xlabel="H2 mole fraction", ylabel="Pressure [Pa]", xlim = (-0.05,1),xticks=(0, 0.2, 0.4, 0.6, 0.8, 1), ylim=(0,400e5), title=f"vdW, Temperature = {temperature} K")
    ax.grid()
    ax.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```

```python
#temperatures to use: 298, 311, 378 (highest number of experimental points)

# classes
from eos import EOS
from eos_mixture import EOSMixture
from  alpha_r_cubic import AlphaRCubic
from mixture_fluid import MixtureFluid

# objects
from purefluid import NH3, H2
from cp0 import NH3_cp0_NIST, H2_cp0_NIST 


# algorithms and other functions and variables
from symbols import *
from mixture_rules import one_fluid_theory
from other_functions import multi_root
from algorithms import VLE_P_dew, VLE_P_bubble, PR_coefficients

# general
import pandas as pd
import matplotlib.pyplot as plt

####################################### Temperature
temperatures = [298, 311, 378]

####################################### Experimental data
path ="../dataframes/ammonia_hydrogen_VLE.xlsx"
data = pd.read_excel(path)

# round temperature to closest kelvin
data["T"] = round(data["T"], 0)


####################################### EOS

#expressions for PR
PR = {
    "alpha_r_expr": sp.log(abs(1/rho/(1/rho-b))) + 1/(R*T)*a/(4*b)*2**0.5*sp.log(abs((1/rho-b*(2**0.5-1))/(1/rho+b*(2**0.5+1)))),
    "a_c_expr": 0.4572*R**2*T_c**2/P_c,
    "alpha_T_expr": (1+(0.37464+1.54226*omega - 0.2699*omega**2)*(1-(T/T_c)**0.5))**2,
    "b_expr": 0.0778*R*T_c/P_c
}

#create pure fluid EOS from PR
PR_NH3 = EOS("PR_NH3", NH3, NH3_cp0_NIST, AlphaRCubic, **PR)
PR_H2 = EOS("PR_H2", H2, H2_cp0_NIST, AlphaRCubic, **PR)

#create mixture and mixture EOS
mix = MixtureFluid(H2, NH3)

mix_EOS = EOSMixture("PR_H2_NH3", mix, AlphaRCubic, PR_H2, PR_NH3, mixture_rule=one_fluid_theory)

####################################### calculations for various temperatures
for temperature in temperatures:
    # restrict data to matching temperature
    data_T = data[data["T"]==temperature]

    # calculate dew compositions and pressure from algorithm
    compositions_dew, pressures_dew = VLE_P_dew(mix_EOS, temperature, method="coefficients", pressure_break=400e5, step = 1e-2, coefficient_function=PR_coefficients)

    # calculate bubble compositions and pressure from algorithm
    compositions_bubble, pressures_bubble = VLE_P_bubble(mix_EOS, temperature, method="coefficients", pressure_break=400e5, step = 0.005, pressure_jump=1e5, coefficient_function=PR_coefficients)


    ####################################### Plotting
    fig, ax = plt.subplots()

    ax.plot(compositions_dew, pressures_dew, label="Dew line")
    ax.plot(compositions_bubble, pressures_bubble, label="Bubble line")

    ax.scatter(data_T["y_H2"], data_T["P"], label="Dew line experimental", color="black", marker="o")
    ax.scatter(data_T["x_H2"], data_T["P"], label="Bubble line experimental", color="black", marker="x")

    ax.set(xlabel="H2 mole fraction", ylabel="Pressure [Pa]", xlim = (-0.05,1),xticks=(0, 0.2, 0.4, 0.6, 0.8, 1), ylim=(0,400e5), title=f"PR, Temperature = {temperature} K")
    ax.grid()
    ax.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```

```python
#temperatures to use: 298, 311, 378 (highest number of experimental points)

# classes
from eos import EOS
from eos_mixture import EOSMixture
from alpha_r_cubic import AlphaRCubic
from alpha_r_helmholtz import AlphaRHelmholtz
from mixture_fluid import MixtureFluid

# objects
from purefluid import NH3, H2
from cp0 import NH3_cp0_NIST, H2_cp0_NIST 

# algorithms and other functions and variables
from symbols import *
from algorithms import VLE_P_dew, VLE_P_bubble, SRK_coefficients
from presets import alpha_r_0300, alpha_r_0313
from mixture_rules import one_fluid_theory

# general
import pandas as pd
import matplotlib.pyplot as plt

####################################### Temperature
temperatures = [273, 311, 378]
# temperatures = [378]
####################################### Experimental data
path ="dataframes/ammonia_hydrogen_VLE.xlsx"
data = pd.read_excel(path)

# round temperature to closest kelvin
data["T"] = round(data["T"], 0)

####################################### Cubic EOS
#expressions for SRK
SRK = {
    "alpha_r_expr": sp.log(abs((1/rho/(1/rho-b)))) + 1/(R*T)*a/b*sp.log(abs(1/rho/(1/rho+b))),
    "a_c_expr": 0.42748*R**2*T_c**2/P_c,
    "alpha_T_expr": (1+ (0.48 + 1.574*omega - 0.176*omega**2)*(1-(T/T_c)**0.5))**2,
    "b_expr": 0.08664*R*T_c/P_c
}

#create pure fluid EOS from PR
SRK_NH3 = EOS("SRK_NH3", NH3, NH3_cp0_NIST, AlphaRCubic, **SRK)
SRK_H2 = EOS("SRK_H2", H2, H2_cp0_NIST, AlphaRCubic, **SRK)

#create mixture and mixture EOS
mix = MixtureFluid(H2, NH3)

cub_EOS = EOSMixture("SRK_H2_NH3", mix, AlphaRCubic, SRK_H2, SRK_NH3, mixture_rule=one_fluid_theory)

####################################### Helmholtz EOS
#create pure fluid EOS for both fluids
H2_0313 = EOS("H2_0313", H2, H2_cp0_NIST, AlphaRHelmholtz, alpha_r_expr=alpha_r_0313())
NH3_0300 = EOS("NH3_0300", NH3, NH3_cp0_NIST, AlphaRHelmholtz, alpha_r_expr=alpha_r_0300())

# create mixture and mixture EOS (hydrogen must be the first fluid)
mix = MixtureFluid(H2, NH3)

# mixture cofficients from [0301]
n_ij = [-3.73558, -7.47092, 1.98413, 1.87191]
t_ij = [1.28, 2.05, 2.6, 3.13]
d_ij = [1, 2, 1, 2]

eta_ij = [0, 0, 0.61, 1.6]
beta_ij = [0, 0, 2.06, 1.74]
gamma_ij = [0, 0, 0.79, 2.1]
epsilon_ij = [0, 0, 0.8, 1.62]

k_pol = 2
k_exp = 0
k_gbs = 2

#binary parameters from [0301]
beta_T = [[1, 1/0.98824],[0.98824, 1]]  # because [0301] take ammonia to be the first component while I use hydrogen
gamma_T = [[1,1.1266], [1.1266,1]]
beta_V = [[1, 1/1.0103],[1.0103, 1]]    # because [0301] take ammonia to be the first component while I use hydrogen
gamma_V = [[1,0.7298], [0.7298,1]]

helm_EOS = EOSMixture("H2_0313_NH3_0300", mix, AlphaRHelmholtz, H2_0313, NH3_0300)#, 
                    #  F=1, n_ij=n_ij, t_ij=t_ij, d_ij=d_ij, eta_ij=eta_ij, beta_ij=beta_ij, gamma_ij=gamma_ij, epsilon_ij=epsilon_ij,
                    #  beta_T=beta_T, beta_V=beta_V, gamma_T=gamma_T, gamma_V=gamma_V,
                    #  k_pol=k_pol, k_gbs=k_gbs)


####################################### calculations for various temperatures
for temperature in temperatures:
    # restrict data to matching temperature
    data_T = data[data["T"]==temperature]

    fig, ax = plt.subplots()

    ####################################### Cubic
    # calculate dew compositions and pressure from algorithm
    compositions_dew, pressures_dew = VLE_P_dew(cub_EOS, temperature, method="coefficients", pressure_break=500e5, step = 5e-3, coefficient_function=SRK_coefficients)

    # calculate bubble compositions and pressure from algorithm
    compositions_bubble, pressures_bubble = VLE_P_bubble(cub_EOS, temperature, method="coefficients", pressure_break=500e5, step = 5e-3,  coefficient_function=SRK_coefficients)

    ax.plot(compositions_dew, pressures_dew, label="Dew line, SRK")
    ax.plot(compositions_bubble, pressures_bubble, label="Bubble line, SRK")

    ####################################### Helmholtz
    # calculate dew compositions and pressure from algorithm
    compositions_dew, pressures_dew = VLE_P_dew(helm_EOS, temperature, method="multi-root", pressure_break=500e5, step = 1e-3, n=1.5e3, pressure_factor=0.6)

    # calculate bubble compositions and pressure from algorithm
    compositions_bubble, pressures_bubble = VLE_P_bubble(helm_EOS, temperature, method="multi-root", pressure_break=500e5, step = 1e-4, n=1.5e3, pressure_factor=0.6)

    ax.plot(compositions_dew, pressures_dew, label="Dew line, [0300] + [0313]")
    ax.plot(compositions_bubble, pressures_bubble, label="Bubble line, [0300] + [0313]")


    ####################################### Plotting settings and experimental data


    ax.scatter(data_T["y_H2"], data_T["P"], label="Dew line experimental", color="black", marker="o")
    ax.scatter(data_T["x_H2"], data_T["P"], label="Bubble line experimental", color="black", marker="x")

    ax.set(xlabel="H2 mole fraction", ylabel="Pressure [Pa]", xlim = (-0.05,1),xticks=(0, 0.2, 0.4, 0.6, 0.8, 1), ylim=(0,500e5), title=f"Temperature = {temperature} K")
    ax.grid()
    ax.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```
